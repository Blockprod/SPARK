## 3. Configurer les secrets

```python
import os, shutil
from kaggle_secrets import UserSecretsClient

secrets = UserSecretsClient()

# -- Secrets requis (erreur explicite si absent) ------------------------------
REQUIRED_SECRETS = ['GEMINI_API_KEY']
# -- Secrets optionnels (ignorés silencieusement si absents) ------------------
OPTIONAL_SECRETS: list[str] = []

env_lines = []
missing_required = []

for name in REQUIRED_SECRETS:
    try:
        value = secrets.get_secret(name)
        env_lines.append(f'{name}={value}')
        os.environ[name] = value          # injecte dans le process ? hérité par !python
        print(f'  ? {name}')
    except Exception:
        missing_required.append(name)
        print(f'  ? {name} — MANQUANT (requis)')

for name in OPTIONAL_SECRETS:
    try:
        value = secrets.get_secret(name)
        env_lines.append(f'{name}={value}')
        os.environ[name] = value
        print(f'  ? {name} (optionnel)')
    except Exception:
        print(f'  - {name} absent (optionnel, ignoré)')

if missing_required:
    raise RuntimeError(
        f"Secrets Kaggle manquants : {missing_required}\n"
        "? Notebook ? Add-ons ? Secrets ? activer le secret (toggle ON)"
    )

with open('/kaggle/working/SPARK/.env', 'w') as f:
    f.write('\n'.join(env_lines))
print(f'\n.env écrit ({len(env_lines)} variables) ?')
print('os.environ mis à jour ? (hérité par les sous-processus)')

# Copie les fichiers secrets depuis le dataset privé spark-secrets
DATASET_DIR = '/kaggle/input/spark-secrets'
os.makedirs('/kaggle/working/SPARK/secrets', exist_ok=True)
os.makedirs('/kaggle/working/SPARK/models', exist_ok=True)

for fname in ['client_secret.json', 'youtube_token.json']:
    src = f'{DATASET_DIR}/{fname}'
    if os.path.exists(src):
        shutil.copy(src, f'/kaggle/working/SPARK/secrets/{fname}')
        print(f'  {fname} ?')
    else:
        print(f'  {fname} absent (upload YouTube désactivé)')

for fname in ['kokoro-v1.0.fp16.onnx', 'voices-v1.0.bin']:
    src = f'{DATASET_DIR}/models/{fname}'
    if os.path.exists(src):
        shutil.copy(src, f'/kaggle/working/SPARK/models/{fname}')
        print(f'  {fname} ?')

print('Secrets configurés ?')
```

## 6. Vérifier la config LTX-Video

```python
import yaml
config_path = '/kaggle/working/SPARK/config.yaml'
with open(config_path, 'r') as f:
    cfg = yaml.safe_load(f)

cfg['video_generation']['provider'] = 'ltx'
cfg['video_generation']['model_id'] = 'Lightricks/LTX-Video'
cfg['video_generation']['device'] = 'cuda'
cfg['video_generation']['num_frames'] = 217      # 8*27+1, ~9.0s/scène a 24fps
cfg['video_generation']['num_inference_steps'] = 25
cfg['video_generation']['guidance_scale'] = 3.0
cfg['video_generation']['use_cpu_offload'] = True
if 'scenes' not in cfg['video_generation']:
    cfg['video_generation']['scenes'] = {}
cfg['video_generation']['scenes']['max_scene_duration_sec'] = 10
cfg['pipeline']['target_width'] = 288            # 9:16 exact, ~4× moins de tokens
cfg['pipeline']['target_height'] = 512
cfg['pipeline']['target_fps'] = 24
cfg['post_production']['subtitles']['style']['font_size'] = 42
cfg['post_production']['subtitles']['style']['bold'] = True
cfg['post_production']['subtitles']['style']['margin_v'] = 70
if 'audio_generation' not in cfg:
    cfg['audio_generation'] = {}
cfg['audio_generation']['active_backend'] = 'edge_tts'
if 'edge_tts' not in cfg['audio_generation']:
    cfg['audio_generation']['edge_tts'] = {}
cfg['audio_generation']['edge_tts']['voice'] = 'fr-FR-VivienneMultilingualNeural'
cfg['audio_generation']['edge_tts']['rate'] = '-5%'
cfg['audio_generation']['edge_tts']['pitch'] = '+0Hz'

with open(config_path, 'w') as f:
    yaml.dump(cfg, f, allow_unicode=True, default_flow_style=False)

with open(config_path, 'r') as f:
    check = yaml.safe_load(f)
v = check['video_generation']
p = check['pipeline']
print(f"  Provider        : {v['provider']}")
print(f"  Model           : {v.get('model_id', 'N/A')}")
print(f"  Resolution      : {p['target_width']}x{p['target_height']}")
print(f"  FPS             : {p['target_fps']}")
print(f"  Num frames      : {v.get('num_frames', 'N/A')} (8k+1)")
print(f"  Inference steps : {v.get('num_inference_steps', 'N/A')}")
print(f"  CPU offload     : {v.get('use_cpu_offload', False)}")
print(f"  Audio backend   : {check['audio_generation']['active_backend']}")
print()
print('Config vérifiée ? — prêt à lancer le pipeline')
```
